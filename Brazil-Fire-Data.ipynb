```python
import os
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
%matplotlib inline
```

```python
map_months={
    'Janeiro':'January',
    'Fevereiro':'Febraury',
    'Marco':'March',
    'Abril':'April',
    'Maio':'May',
    'Junho':'June',
    'Julho':'July',
    'Agosto':'August',
    'Setembro':'September',
    'Outurbo':'October',
    'Novembero':'November',
    'Dezembro':'December'
}
```

```python
file=pd.read_csv(r"C:\Users\KANDEEBARAJ\Downloads\amazon.csv" ,encoding='cp1252')
file=file.drop_duplicates()
file=file.replace(map_months)
```

```python
file
```

```python
file.columns
```

```python
avg_fires_month=file.groupby('month',sort=False)['number'].mean().reset_index()
```

```python
plt.plot(avg_fires_month['month'],avg_fires_month['number'],marker='s')
plt.xlabel('Month')
plt.ylabel('Avg no of fires')
plt.gca().tick_params(axis='x',rotation=90)
```

```python
plt.bar(avg_fires_month['month'],avg_fires_month['number'],color='g')
plt.xlabel('Month',weight='bold')
plt.ylabel('Avg. no. fires',weight='bold')
plt.gca().tick_params(axis='x',rotation=90)
```

```python
state_avg_fire=file.groupby('state')['number'].mean().sort_values(ascending=False).reset_index()
```

```python
plt.barh(state_avg_fire['state'],state_avg_fire['number'])
plt.xlabel('Avg. no. Fires',weight='bold')
plt.ylabel('State',weight='')
plt.gca().invert_yaxis()
```

```python
year_avg_fires=file.groupby(['year'])['number'].sum().reset_index()
```

```python
plt.bar(year_avg_fires['year'],year_avg_fires['number'])
plt.xlabel('Year',weight='bold')
plt.ylabel('No. of Avg. Fires',weight='bold')
plt.xticks(np.arange(1998,2018,1))
plt.gca().tick_params(axis='x',rotation=90)
```
